```python
# itterrows() : 데이터프레임에서 행에 반복 접근을 하며 값을 추출/ 조작할 때 사용
```

```python
import pandas as pd
import io
import requests
import pprint

url = 'https://raw.githubusercontent.com/chloevan/datasets/master/sports/baseball_stats.csv'
url=requests.get(url).content
baseball_stats = pd.read_csv(io.StringIO(url.decode('utf-8')))

baseball_stats.head()
#pprint.pprint(baseball_stats.head())
```

```python
bos_df=baseball_stats[ ( baseball_stats.Team=="BOS" ) & ( baseball_stats.Year.isin([2008,2009,2010]) ) ].reset_index(drop=True)   # drop=True로 하면 인덱스가 다시 0부터 시작
bos_df[:5]
```

```python
for i, row in bos_df.iterrows():   # row의 type은 seies
  print(i)
  print(row)
```

```python
for row in bos_df.iterrows():      # row의 type은 tuple
  print(row)
```

```python
# 득점, 실점 계산

def calc_diff(rs,ra):
  return rs-ra

diffs=[]
for i,row in bos_df.iterrows():
  rs=bos_df["RS"]
  ra=bos_df["RA"]
  run_diff=calc_diff(rs,ra)
  diffs.append(run_diff)

bos_df['RD']=diffs
bos_df[:5]

```
